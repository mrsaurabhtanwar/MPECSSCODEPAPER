# MPECSS - MacMPEC Parameter Sensitivity (t0 = 0.1)

This notebook runs one parameter-sensitivity point on the 191 MacMPEC problems.

- Timeout per problem: 1800 s
- Baseline seed: 42
- Sweep point: t0 = 0.1
- Output root: /kaggle/working/outputs_macmpec_param_t0_0p1

Run the setup cell first. After a Kaggle restart, re-run setup, then use Resume instead of Fresh Run.


```python
import os
from pathlib import Path
import shutil
import subprocess
import sys

DATASET = 'macmpec'
RUN_TAG_PREFIX = 'Kaggle_MacMPEC_ParamSensitivity_t0_0p1'
WORKERS = 4
TIMEOUT = 1800
DEFAULT_SEED = 42
SAVE_LOGS = True
SHUFFLE = True

REPO_DIR = '/kaggle/working/MPECSSCODEPAPER'
REPO_GIT_URL = 'https://github.com/mrsaurabhtanwar/MPECSSCODEPAPER.git'
OUTPUT_ROOT = '/kaggle/working/outputs_macmpec_param_t0_0p1'
DATASET_PATH = os.path.join(REPO_DIR, 'benchmarks', 'macmpec', 'macmpec-json')
REPO_COMMIT = 'SET_RELEASE_COMMIT'

RUN_PLAN = [
    {
        "slug": "t0-0p1",
        "tag": "Kaggle_MacMPEC_ParamSensitivity_t0_0p1",
        "solver_params": {
            "t0": 0.1
        }
    }
]

repo_path = Path(REPO_DIR)
if repo_path.exists():
    shutil.rmtree(repo_path)

if REPO_COMMIT == 'SET_RELEASE_COMMIT':
    raise RuntimeError('Set REPO_COMMIT to the release commit before the final Kaggle run.')

print(f'Cloning: {REPO_GIT_URL} @ {REPO_COMMIT}')
subprocess.run(['git', 'clone', '--depth', '1', REPO_GIT_URL, str(repo_path)], check=True)
subprocess.run(['git', 'checkout', REPO_COMMIT], check=True, cwd=str(repo_path))
sys.path.insert(0, str(repo_path))

subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', '--upgrade', 'pip'], check=True)
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', '-e', str(repo_path)], check=True)

print(f'[OK] Repo ready: {repo_path}')
print(f'[OK] Benchmark path: {DATASET_PATH}')
if not os.path.exists(DATASET_PATH):
    raise RuntimeError(f'Dataset path not found: {DATASET_PATH}')
problem_count = len([name for name in os.listdir(DATASET_PATH) if name.endswith('.json')])
print(f'[OK] MacMPEC problems found: {problem_count}')

subprocess.run([sys.executable, os.path.join(str(repo_path), 'scripts', 'preflight_checks.py')], check=True)
```

```python
from kaggle_setup.study_runner import run_study_plan

def run_fresh():
    run_study_plan(
        RUN_PLAN,
        repo_dir=REPO_DIR,
        dataset=DATASET,
        path=DATASET_PATH,
        output_root=OUTPUT_ROOT,
        workers=WORKERS,
        timeout=TIMEOUT,
        seed=DEFAULT_SEED,
        save_logs=SAVE_LOGS,
        shuffle=SHUFFLE,
    )

def run_resume():
    run_study_plan(
        RUN_PLAN,
        repo_dir=REPO_DIR,
        dataset=DATASET,
        path=DATASET_PATH,
        output_root=OUTPUT_ROOT,
        workers=WORKERS,
        timeout=TIMEOUT,
        seed=DEFAULT_SEED,
        save_logs=SAVE_LOGS,
        shuffle=SHUFFLE,
        resume_latest=True,
    )

def run_summary():
    run_study_plan(
        RUN_PLAN,
        repo_dir=REPO_DIR,
        dataset=DATASET,
        path=DATASET_PATH,
        output_root=OUTPUT_ROOT,
        workers=WORKERS,
        timeout=TIMEOUT,
        seed=DEFAULT_SEED,
        save_logs=SAVE_LOGS,
        shuffle=SHUFFLE,
        summary_only=True,
    )
```

```python
run_fresh()
```

```python
run_resume()
```

```python
run_summary()
```
